```python
import streamlit as st
import pandas as pd
import plotly.express as px
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
data_import = pd.read_csv("MPG_data.csv")
data_import.head()
```

```python
numerical_colunms_list = []
categorical_columns_list = []

for i in data_import.columns:
    if data_import[i].dtype == np.dtype("float64") or data_import[i].dtype == np.dtype("int64"):
        numerical_colunms_list.append(data_import[i])
    else:
        categorical_columns_list.append(data_import[i])
```

```python
numerical_data = pd.concat(numerical_colunms_list, axis=1)
numerical_data.head()
```

```python
categorical_data = pd.concat(categorical_columns_list, axis=1)
categorical_data.head()
```

```python
numerical_data.apply(lambda x: x.fillna(np.mean(x)))
```

```python
mod_numerical_data = numerical_data.dropna(axis=1)
mod_numerical_data.head()
```

```python
mod_numerical_data.isnull().sum()
```

```python
scaler = StandardScaler()

scaled_values = scaler.fit_transform(mod_numerical_data)
scaled_values
```

```python
pca = PCA()

pca_data = pca.fit_transform(scaled_values)

pca_data = pd.DataFrame(pca_data)
pca_data
```

```python
new_column_names = ["PCA_"+str(i) for i in range(1, len(pca_data.columns)+1)]
```

```python
list(pca_data.columns)
column_mapper = dict(zip(list(pca_data.columns), new_column_names))
```

```python
# Chaning colunms names 
pca_data = pca_data.rename(columns=column_mapper)
pca_data.head()
```

```python
output = pd.concat([data_import, pca_data], axis=1)
output.head()
```
